```python
import numpy as np
from scipy.signal import butter, lfilter, freqz
import matplotlib.pyplot as plt


def butter_lowpass(cutoff, fs, order=5):
    nyq = 0.5 * fs
    normal_cutoff = cutoff / nyq
    b, a = butter(order, normal_cutoff, btype='low', analog=False)
    return b, a

def butter_lowpass_filter(data, cutoff, fs, order=0.2):
    b, a = butter_lowpass(cutoff, fs, order=order)
    y = lfilter(b, a, data)
    return y


def butter_lowpass_filter_2(data, cutoff, fs, order=4):
    b, a = butter_lowpass(cutoff, fs, order=order)
    y = lfilter(b, a, data)
    return y

# Filter requirements.
order = 5
fs = 200.0       # sample rate, Hz
cutoff = 6.667  # desired cutoff frequency of the filter, Hz

# Get the filter coefficients so we can check its frequency response.
b, a = butter_lowpass(cutoff, fs, order)


y = butter_lowpass_filter(lamp_means, cutoff, fs, order)


y = butter_lowpass_filter(lamp_means, cutoff, fs, order)
error= butter_lowpass_filter(lamp_error, cutoff, fs, order)

fig = plt.figure(figsize=(8,4))
fig.set_facecolor('w')

ax = fig.add_subplot(212)
#plt.subplot(2, 1, 2)
ax.plot(t, lamp_means, 'b-', label='data')
ax.plot(t, y, 'g-', linewidth=2, label='filtered data')
#ax.xlabel('Time [sec]')
plt.grid()
plt.legend()
ax.set_ylim(475, 525)

plt.subplots_adjust(hspace=0.35)
```
